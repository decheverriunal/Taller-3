```python
import random
import numpy as np

ITER = 5000
SAMPLE = 200

# numero de curules de cada partido
listaCurules = np.array([8,8,16,13,5])

# nombre de cada entidad
nombreEntidades = ["Biblioteca central","Teatro mayor","Hospital central",
                   "Estadio","Hospital principal","Plaza central",
                   "Iglesia grande","Biblioteca del norte","Iglesia pequeña",
                   "Centro comercial","Centro financiero","Palacio del gobierno",
                   "Colegio distrital","Universidad distrital",
                   "Iglesia principal","Banco central","Superintendencia",
                   "Ministerio de defensa","Corte suprema",
                   "Instituto investigativo","Aeropuerto","Banco nacional",
                   "Centro de estudios","Centro de investigación",
                   "Dirección general","Cámara","Agencia tributaria",
                   "Fondo monetario","Instituto central","Administración federal",
                   "Ministerio de defensa","Comisión nacional",
                   "Instituto de ciencias","Policía nacional",
                   "Comisión reguladora","Ministerio de vivienda",
                   "Corte constitucional","Secretaria de cultura",
                   "Banco de desarrollo","Agencia ambiental","Ministerio publico",
                   "Secretaria de economía","Ministerio de turismo","Gran teatro",
                   "Dirección de infraestructura","Consejo de administración",
                   "Instituto estadístico","Agencia de transporte",
                   "Agencia de migración","Agencia de regulación"]

# peso politico de cada entidad
entidades = np.array([71,20,63,5,83,46,49,37,2,26,
             49,99,85,46,83,37,15,40,40,13,
             84,48,97,64,15,38,50,49,19,59,
             39,73,52,4,57,24,75,98,5,25,
             39,48,91,32,77,81,4,21,11,63])

# generar lista de cromosomas
# cada 3 bits representan una entidad e indican a que partido pertenece
cList = np.zeros((SAMPLE,150))
for i in range(SAMPLE):
  for j in range(150):
    if random.random() > 0.5:
      cList[i][j] = 1

# evaluacion
apt = np.zeros(SAMPLE)
def eval(list):
  c = len(list[0])//3
  curules = np.zeros(5)
  for i in range(len(list)):
    curules[:] = listaCurules
    for j in range(c):
      n = int(list[i][j*3]*4+list[i][j*3+1]*2+list[i][j*3+2])
      if n == 0 and curules[0] > 0:
        apt[i] += entidades[j]
      if n <= 4:
        curules[n] -= 1
      else:
        apt[i] -= 25
    for j in range(len(curules)):
      if curules[j] < 0:
        apt[i] -= abs(curules[j]*20)
    apt[i] = int(max(apt[i],1))

# probabilidades
probabilityList = np.zeros(SAMPLE)
def prob():
  t = 0
  for i in range(SAMPLE):
    t += apt[i]
  for i in range(SAMPLE):
    if t != 0:
      probabilityList[i] = apt[i] / t

# seleccion de padres
parents = np.zeros((SAMPLE,150))
def getParents():
  for i in range(SAMPLE):
    total = 0
    r = random.random()
    for j in range(SAMPLE):
      total += probabilityList[j]
      if r < total:
        parents[i][:] = cList[j]
        break

# cruce
def getChildren():
  for i in range(0,SAMPLE-1,2):
    r = random.randrange(len(cList[0])//3-1)+1
    a = np.zeros(len(cList[0])-r)
    a[:] = parents[i][r:len(parents[0])]
    cList[i][:] = np.concatenate([parents[i][0:r],parents[i+1][r:len(parents[0])]])
    cList[i+1][:] = np.concatenate([parents[i+1][0:r],a])

# mutacion
def mut():
  ri = random.randrange(len(cList))
  rj = random.randrange(len(cList[0]))
  if cList[ri][rj] == 0:
    cList[ri][rj] = 1
  else:
    cList[ri][rj] = 0

for i in range(ITER):
  apt = np.zeros(SAMPLE, dtype=int)
  eval(cList)
  if (i+1) % 100 == 0:
    print("Generacion " + str(i+1))
    print("Puntaje promedio: " + str(np.average(apt)))
    print("Mejor puntaje: " + str(np.max(apt)))
    print("-----------------")
  prob()
  getParents()
  getChildren()
  for i in range(10):
    mut()

apt = np.zeros(SAMPLE, dtype=int)
eval(cList)
print("\nResultados: \n")
print("Puntaje promedio de ultima generacion: " + str(np.average(apt)))
print("Puntaje de mejor cromosoma: " + str(np.max(apt)))
print("Curules asignados al partido 1:\n")
p = [0] * (len(cList[0])//3)
indL, = np.where(apt == np.max(apt))
ind = indL[0]
for i in range(len(p)):
  p[i] = int(4*cList[ind][i*3] + 2*cList[ind][i*3+1] + cList[ind][i*3+2])
print("Puntos\t|\tNombre de entidad")
print("---------------------------------------------")
for i in range(len(p)):
  if p[i] == 0:
    print(str(entidades[i]) + "\t|\t" + nombreEntidades[i])

#for i in range(SAMPLE):
#  p = [0] * (len(cList[0])//3)
#  for j in range(len(p)):
#    p[j] = int(4*cList[i][j*3] + 2*cList[i][j*3+1] + cList[i][j*3+2])
#  print(p)
```
